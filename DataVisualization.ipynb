### Sample size and names
-> N=1
    case study = one patient
    anecdote = one person
    (interesting stories that inspire future research but that should not be trusted or overly generalized)

-> N > 1(different names baed on sample size)
    pilot study
    proof of principle
    small sample case study
    large sample case stude
    (actual researches with potential importance and generalized findings)

### Pie Chart in Matplotlib

```python
preferences = [5, 40, 35, 20]
explode = [0, 0, .05, .15]
plt.pie(preferences, explode=explode)
plt.show()
```

```python
preferences = [5, 40, 35, 20]
explode = [0, 0, .05, .15]
plt.pie(preferences, explode=explode, autopct='%1.f%%')
plt.show()
```

```python
preferences = [5, 40, 35, 20]
ice_cream_flavors = [ 'Fried chicken', 'chocolate', 'Vanilla','Green Tea']

# explosion factor
explode = [0, 0, .05, .15]

plt.pie(preferences, labels=ice_cream_flavors,
       explode=explode, shadow=True, autopct='%1.f%%')
plt.show()
```

### Histograms in Matplotlib



```python
N = 500
data = np.exp(np.random.randn(N) + 2)
data
```

```python
plt.figure(figsize=(10, 5))
```

```python
plt.hist(data, bins=40)
```

```python
plt.hist(data, bins=40, density=True)
```

```python
plt.hist(data, bins="fd", density=True)
# bins='fd' uses the FreedmanDiaconis rule to choose the *bin width*;
# the number of bins is then derived from that width.
plt.title("Histogram of Data")
plt.xlabel("Data value")
plt.ylabel("Bin count")
plt.show()
```

```python
d1 = np.random.randn(N) + 4
d2 = np.random.randn(N)
d3 = np.random.randn(N) + 2
```

```python
np.histogram(d1, bins=42)
```

```python
y1, x1 = np.histogram(d1, bins=42)
y2, x2 = np.histogram(d2, bins=42)
y3, x3 = np.histogram(d3, bins=42)

```

```python
y1
```

```python
100*y1/np.sum(y1)
```

```python
y1 = 100*y1/np.sum(y1)
y2 = 100*y2/np.sum(y2)
y3 = 100*y3/np.sum(y3)
```

```python
plt.figure(figsize=(10,5))
```

```python
plt.plot(x1,y1)
```

```python
plt.plot(x1[1:], y1)
```

```python
plt.plot((x1[:-1]+x1[1:])/2, y1)
```

```python
plt.plot((x1[:-1]+x1[1:])/2, y1, 's-')
```

```python
plt.plot((x1[:-1]+x1[1:])/2, y1, 'o-')
```

```python
plt.plot((x1[:-1]+x1[1:])/2, y1, 's-', label='Data 1')
```

```python
plt.figure(figsize=(10,5))
plt.plot((x1[:-1]+x1[1:])/2, y1, 's-', label='Data 1')
plt.plot((x2[:-1]+x2[1:])/2, y2, 'o-', label='Data 2')
plt.plot((x3[:-1]+x3[1:])/2, y3, 'o-', label='Data 3')

plt.xlabel("Data value")
plt.ylabel("Percent of total")

plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot((x1[:-1]+x1[1:])/2, y1, 's-', label='Data 1')
plt.plot((x2[:-1]+x2[1:])/2, y2, 'o-', label='Data 2')
plt.plot((x3[:-1]+x3[1:])/2, y3, 'o-', label='Data 3')

plt.xlabel("Data value")
plt.ylabel("Percent of total")
plt.legend()
plt.show()
```

```python
plt.hist(d1, bins="fd", density=True)
plt.hist(d2, bins="fd", density=True)
plt.hist(d3, bins="fd", density=True)
# bins='fd' uses the FreedmanDiaconis rule to choose the *bin width*;
# the number of bins is then derived from that width.
plt.title("Histogram of Data")
plt.xlabel("Data value")
plt.ylabel("Bin count")
plt.show()
```
