```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = [[10, 12, 96, 40], [90, 20, 55, 15]]

sources = ['TV', 'Newspaper', 'Internet', 'Word of Mouth']
groups = ['20-40', '50-70']
```

```python
zip(data)
```

```python
list(zip(data))
```

```python
list(zip(*data))
```

```python
df = pd.DataFrame(zip(*data), index=sources, columns=groups)
df
```

```python
df.plot.bar()
plt.show()
```

```python
df.T.plot.bar()
plt.show()
```

```python
df.T.plot.bar(figsize=(8,4), ylabel="How often (parent)",
             title="How do people get the news", fontsize=14)
plt.show()
```

### Pie Chart in Matplotlib

```python
preferences = [5, 40, 35, 20]
plt.pie(preferences)
plt.show()
```

```python
preferences = [5, 40, 35, 20]
explode = [0, 0, .05, .15]
plt.pie(preferences, explode=explode)
plt.show()
```

```python
preferences = [5, 40, 35, 20]
explode = [0, 0, .05, .15]
plt.pie(preferences, explode=explode, autopct='%1.f%%')
plt.show()
```

```python
preferences = [5, 40, 35, 20]
ice_cream_flavors = [ 'Fried chicken', 'chocolate', 'Vanilla','Green Tea']

# explosion factor
explode = [0, 0, .05, .15]

plt.pie(preferences, labels=ice_cream_flavors,
       explode=explode, shadow=True, autopct='%1.f%%')
plt.show()
```

### Box plots in seaborn

```python
N = 200

```

```python

# uniformly distributed random numbers
y1 = np.random.rand(N)
#normally distributed(galsian distributed) random numbers
y2 = np.random.randn(N)
```

```python
y1
```

```python
y2
```

```python
sns.boxplot(data=[y1, y2])
plt.show()
```

```python
sns.boxplot(data=[y1, y2], palette='Purples')
plt.show()
```

```python
sns.stripplot(data=[y1, y2])
plt.show()
```

```python
sns.stripplot(data=[y1, y2], marker='.')
plt.show()
```

```python
sns.stripplot(data=[y1, y2], marker='.', color="k")
plt.show()
```

```python
sns.boxplot(data=[y1, y2], palette='Purples')
sns.stripplot(data=[y1, y2], marker='.', color="k")
plt.show()
```

```python
# by importing to pandas
data = np.stack((y1, y2), axis=1)
df = pd.DataFrame(data, columns=['var1', 'var2'])

df.boxplot()
```
